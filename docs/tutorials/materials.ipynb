# Material Database Query

OpEnUV bundles the CXRO/Henke scattering-factor database for 92 elements.
This tutorial shows how to query refractive indices.

```python
import sys, os
sys.path.insert(0, os.path.abspath('../src'))

import numpy as np
import matplotlib.pyplot as plt

from euv.materials import CXROTable
from euv.constants import EV_TO_JOULE, PLANCK, SPEED_OF_LIGHT
```

```python
table = CXROTable()
available = table.available_elements()
print(f"Available elements: {len(available)}")
print(available[:20])
```

```python
# Mo/Si multilayer materials at 13.5 nm (91.84 eV)
for el in ['Mo', 'Si', 'Ru', 'Ta', 'Te']:
    n, k = table.refractive_index(el, 91.84)
    print(f"{el:>4s}: n = {n:.6f}, k = {k:.6f}, δ = {1-n:.6f}")
```

```python
# Energy sweep for a single element
energies = np.linspace(50, 200, 500)
ns, ks = [], []
for e in energies:
    try:
        n, k = table.refractive_index('Si', e)
        ns.append(n); ks.append(k)
    except (ValueError, FileNotFoundError):
        ns.append(np.nan); ks.append(np.nan)

wavelengths_nm = [
    (PLANCK * SPEED_OF_LIGHT) / (e * EV_TO_JOULE) * 1e9 for e in energies
]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(wavelengths_nm, ns)
ax1.set_xlabel('Wavelength [nm]')
ax1.set_ylabel('n (refractive index)')
ax1.set_title('Si refractive index')
ax1.axvline(13.5, color='r', ls='--', alpha=0.5, label='13.5 nm')
ax1.legend()

ax2.plot(wavelengths_nm, ks)
ax2.set_xlabel('Wavelength [nm]')
ax2.set_ylabel('k (extinction)')
ax2.set_title('Si extinction coefficient')
ax2.axvline(13.5, color='r', ls='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
